```python
import numpy as np
import torch
from scipy.linalg import toeplitz
from MCMC.BayesLogit import Bayes_Logit
from Optimization.CD import SLR
import matplotlib.pyplot as plt
```

```python
N=600
P=200
beta=np.zeros((P,1))
Corr=0.5**toeplitz(np.arange(0,P),np.arange(0,P))
beta[0]=3
beta[1]=1.5
beta[4]=2
beta[9]=1
beta[12]=1
beta[18]=0.5
beta[25]=-0.5
beta[30]=2.0
beta[45]=-1.2
beta[50]=-1
X=np.random.multivariate_normal(np.zeros(P), Corr, N)
Z=X@beta
Probability=np.exp(Z)/(1+np.exp(Z))
Y=np.random.binomial(1, Probability)
```

```python
device1 = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# Simulation
beta_sample=Bayes_Logit(torch.from_numpy(Y).to(device1),torch.from_numpy(X).to(device1),1,0.05*P,1)
```

```python
#Optimisation
beta_estimator,beta_path=SLR(torch.from_numpy(Y).to(device1),torch.from_numpy(X).to(device1),C=0.5,s=1,path=True)
```

```python
beta_mean=beta_sample.mean(axis=0)
```

```python
plt.figure(figsize=(12, 5))
plt.rcParams['font.size'] = 10

plt.plot(beta_sample[:,[0, 1, 4, 9,12,18,25,30,45,50]].cpu().numpy())
plt.xlabel('MCMC iteration')
plt.ylabel(r'$\beta$', rotation=0, labelpad=10)
plt.show()
```
